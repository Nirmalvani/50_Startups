```python
plt.figure(figsize=(50,50))
plotnumber = 1

for col in df:
    if plotnumber <= 7 :
        ax = plt.subplot(3,3,plotnumber)
        sns.histplot(x= df[col],kde="True")
        plt.xlabel(col,fontsize=35)
        plt.ylabel("Count",fontsize=35)
        plt.xticks(fontsize=25)
        plt.yticks(fontsize=25)
    plotnumber+=1
plt.show()
```

#### Univarate analysis

- Adminstration --> More than 12 company's have spent around 120000 on adminstration.
- Marketing spend ---> Most company's have spent between 300000 to 200000 on marketing and only few have spent over 400000 on marketing.
- State - Each state have around 16 start ups

```python
X = df.iloc[:,0:-1]
X.sample(1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_new,y,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,classification_report
```

```python
mean = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
```
